```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

生成lena图片不同空间分辨率的图片：
原图分辨率为512*512，输出从左往右、从上往下四个图img、img_2、img_3、img_4分别对应原图灰度图，降采样1/2、1/4、1/8之后得到的灰度图。

```python
def down_sample(img):
    height, width = img.shape[:2]
    new_img = np.zeros((height//2, width//2))
    new_img = img[::2, ::2]
    return new_img

img = cv2.imread('lena.jpg', 0)

img_2 = down_sample(img)
img_4 = down_sample(img_2)
img_8 = down_sample(img_4)

plt.figure(figsize=(10, 10))
plt.subplot(221), plt.imshow(img, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(img_2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(img_4, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(img_8, 'gray'), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

生成lena图片不同灰度分辨率的图片：
输出的灰度图从左往右、从上往下依次为img2、img4、img8、img16、img32、img64、img128、img256，对应的灰度级分别为2、4、8、16、32、64、128、256。

```python
img256 = cv2.imread('lena.jpg', 0)

def gray_transfer(gray_level,img):
    height, width = img.shape[:2]
    new_img = np.zeros((height, width))
    a = int(255 / gray_level)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            new_img[i][j] = int(img[i][j] / a)
    
    return new_img

img2 = gray_transfer(2,img256)
img4 = gray_transfer(4,img256)
img8 = gray_transfer(8,img256)
img16 = gray_transfer(16,img256)
img32 = gray_transfer(32,img256)
img64 = gray_transfer(64,img256)
img128 = gray_transfer(128,img256)


plt.figure(figsize=(10, 18))

plt.subplot(421), plt.imshow(img2, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(422), plt.imshow(img4, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(423), plt.imshow(img8, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(424), plt.imshow(img16, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(425), plt.imshow(img32, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(426), plt.imshow(img64, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(427), plt.imshow(img128, 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(428), plt.imshow(img256, 'gray'), plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.show()
```
